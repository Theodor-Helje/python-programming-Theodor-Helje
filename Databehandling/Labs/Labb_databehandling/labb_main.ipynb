### g: use df.plot.hist()

### h: continued exploration

The previous histograms made me interested in the correlation between the schools that get the most applications and the schools that accept the most students. I want to see if there is a correlation between applications recieved and applications accepted. <br>
I suspect that the schools getting the most applications are the prestigious schools and not the schools that accept the most applications.<br>
To investigate further i will start by sorting both the universities with the most accepted and the most applications in a descending order and placing them side by side in a DataFrame.

```python
pd.DataFrame({
    'Applications (descending)': college.sort_values(by='Apps', ascending=False).index,
    'Accepted (descending)': college.sort_values(by='Accept', ascending=False).index
})
```

This data clearly shows that my initial guess that the schools that accept the most students werent the same schools as the ones getting the most applications was wrong. It seems that there is at least a very strong correlation between the amount of applications and the amount of accepted applications. I will of course have to show this statistically and not just guess.

```python
college[['Apps', 'Accept']].corr()
```

There is a very strong correlation between applications and accepted applications.

```python
plt.figure(figsize=(12,12))
sns.heatmap(college.select_dtypes(include='number').corr(), annot=True)
```
