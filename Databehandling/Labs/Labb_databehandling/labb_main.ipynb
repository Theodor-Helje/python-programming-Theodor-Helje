## VG-uppgfter: f -> h

### f: create elite variable

Created 'Elite' and used df.value_counts()

```python
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
college.value_counts('Elite')
```

Printed boxplot of 'Outstate' versus 'Elite'

```python
college.boxplot('Outstate', 'Elite')
```

### g: use df.plot.hist()

```python
fig, axes = plt.subplots(2, 3, figsize=(9.5,7))

college['Grad.Rate'].plot.hist(bins=15, ax=axes[0,0], title="Graduation rate: bins=15")
college['Grad.Rate'].plot.hist(bins=100, ax=axes[1,0], title="Graduation rate: bins=100")

college['Apps'].plot.hist(bins=35, ax=axes[0,1], title="Applications: bins=35")
college['Apps'].plot.hist(bins=100, ax=axes[1,1], title="Applications: bins=100")

college['Accept'].plot.hist(bins=35, ax=axes[0,2], title="Accepted apps: bins=35")
college['Accept'].plot.hist(bins=100, ax=axes[1,2], title="Accepted apps: bins=100")
```

Conclutions: <br>
The graduation rate follows a normal distribution between the colleges in the data.<br>
There is one outlier with a graduationrate of over 100%.<br>
The applications show that there are only very few schools that get an above average amount of applications<br>
The accepted applications show that across all schools, the amount of accepted applications seem proportional to the amount of applications.<br>
(this does not, however, tell us anything about if the schools getting the most applications are the schools that accept the most students)
